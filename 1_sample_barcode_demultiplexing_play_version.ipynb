# Provide BC context

<div class="alert alert-block alert-success">
    The format should be context-NNN-context, where <b>NNN</b> stands for the sample barcode and <b>context</b> stands for the 3' or 5' DNA sequence around the barcode. Only 3' or 5' context can be used.
</div>

```python
bc_context = 'NNNNTCGAAT'
```

# Get the reference BC:sequence table

<div class="alert alert-block alert-success">
    This table [<i>bc_vs_sq.csv</i>] was exported from the Google Sheets document in a csv format. 
</div>

```python
barcodes_table = pd.read_csv('HTP0006_bc_vs_sq.csv', header=2)
barcodes_table
```

```python
for sq in barcodes_table['Barcode sequence (fixed - NNNN barcode - fixed)']:
    print(sq)
```

<div class="alert alert-block alert-danger">
    The last capital TC letters are the reason for the hardcoded [:-2] here, when extracting the barcode.  
</div>

```python
barcodes_table['barcodes'] = [extract_capital(sq)[:-2] for sq in barcodes_table['Barcode sequence (fixed - NNNN barcode - fixed)']]
```

```python
barcodes_table = barcodes_table[['barcodes', 'Reference Sequence', 'Sequencing ID']]
barcodes_table
```

<div class="alert alert-block alert-success">
    The start and the end of the target sequence are extracted here to trim all the paired reads later on.
</div>

```python
barcodes_table['sequence_start'] = [sq[:7] for sq in barcodes_table['Reference Sequence']]
barcodes_table['sequence_end'] = [sq[-7:] for sq in barcodes_table['Reference Sequence']]
barcodes_table
```

# Read in the NGS samples

```python
#ask Katya

with open('miseq_g008_600_s326_lib53_htp0006_48candidates_1out2_merged_merged.fastq') as f:
    lines = f.readlines()
    
sequences = [lines[i+1].rstrip('\n') for i in range(0,len(lines),4)]
lines
```

# Extract barcodes, demultiplex and trim sequences

```python
demultiplexed_sequences = {}

for sq in sequences:
    bc, new_sq = find_barcode(sq, bc_context)
    if bc in list(HTP006_set1_barcodes_table.barcodes):
        start = HTP006_set1_barcodes_table.sequence_start[HTP006_set1_barcodes_table.barcodes == bc].values[0]
        end = HTP006_set1_barcodes_table.sequence_end[HTP006_set1_barcodes_table.barcodes == bc].values[0]
        try:
            new_sq_trimmed = new_sq[new_sq.index(start):new_sq.index(end)+len(end)]
            if bc in demultiplexed_sequences:
                demultiplexed_sequences[bc].append(new_sq_trimmed)
            else:
                demultiplexed_sequences[bc] = [new_sq_trimmed]
        except ValueError:
            pass
        
```

```python
demultiplexed_sequences = {bc:Counter(demultiplexed_sequences[bc]) for bc in demultiplexed_sequences if len(demultiplexed_sequences[bc])>20}
print(demultiplexed_sequences)
```

# Threshold coverage

```python
coverage_threshold = 10
```

```python
thresholded_demultiplexed_sequences = {}

for bc in demultiplexed_sequences:
    thresholded_demultiplexed_sequences[bc] = {k:v for k, v in demultiplexed_sequences[bc].items() if v > coverage_threshold}

print(thresholded_demultiplexed_sequences)
```

# Extract top sequences

## Dominant sequences

```python
for bc in list(HTP006_set1_barcodes_table.barcodes):
    expected = HTP006_set1_barcodes_table[HTP006_set1_barcodes_table.barcodes==bc]['Reference Sequence'].values[0]
    try:
        all_sqs = sorted(((v,k) for k,v in thresholded_demultiplexed_sequences[bc].items()), reverse=True)
        sq = all_sqs[0][1]
        highlight_differences(sq, expected)
    except:
        print('NO SEQUENCE FOUND')
    
        
```
